```python
# Importing the necessary libraries
import numpy as np
import pandas as pd# importing all the important libraries
import re
import numpy as np
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer
import string
import nltk
import re

from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
sw = stopwords.words('english')
```

```python
# read the positvie text data
pos_rev = pd.read_csv('pos.txt' , sep='\n' , header = None , encoding = 'latin-1')

# adding a tareget column
pos_rev['mood'] = 1.0
pos_rev = pos_rev.rename(columns = {0:'review'})
pos_rev
```

```python
# read the negative text data
neg_rev = pd.read_csv('negative.txt' , sep='\n' , header = None , encoding = 'latin-1')

# adding a tareget column
neg_rev['mood'] = 0.0
neg_rev = neg_rev.rename(columns = {0:'review'})
neg_rev
```

```python
# Cleaning the data

# 1. lower
# 2. remove spaces
# 3. remove punctuation
# 4. remove stopwords
# 5. lemmatize
```

```python
pos_rev.loc[: , 'review'] = pos_rev.loc[: , 'review'].apply(lambda x : x.lower())
pos_rev.loc[: , 'review'] = pos_rev.loc[: , 'review'].apply(lambda x : re.sub(r"@\S+" , "" , x))
pos_rev.loc[: , 'review'] = pos_rev.loc[: , 'review'].apply\
(lambda x : x.translate(str.maketrans(dict.fromkeys(string.punctuation))))
pos_rev.loc[: , 'review'] = pos_rev.loc[: , 'review'].apply\
(lambda x : " ".join([word for word in x.split() if word not in (sw)]))
```

```python
neg_rev.loc[: , 'review'] = neg_rev.loc[: , 'review'].apply(lambda x : x.lower())
neg_rev.loc[: , 'review'] = neg_rev.loc[: , 'review'].apply(lambda x : re.sub(r"@\S+" , "" , x))
neg_rev.loc[: , 'review'] = neg_rev.loc[: , 'review'].apply\
(lambda x : x.translate(str.maketrans(dict.fromkeys(string.punctuation))))
neg_rev.loc[: , 'review'] = neg_rev.loc[: , 'review'].apply\
(lambda x : " ".join([word for word in x.split() if word not in (sw)]))
```

```python
# lemmatization
# pos_rev
from textblob import TextBlob
from textblob import Word

lemmatized_review = []
for i in pos_rev['review']:
    text = TextBlob(i)
    lemmatize=  []
    for j in text.words:
        word = Word(j)
        lemmatize.append(word.lemmatize())
    lemmatized_review.append(" ".join(lemmatize))
pos_rev['lemmatized_review'] = lemmatized_review      
```

```python
lemmatized_review = []
for i in neg_rev['review']:
    text = TextBlob(i)
    lemmatize=  []
    for j in text.words:
        word = Word(j)
        lemmatize.append(word.lemmatize())
    lemmatized_review.append(" ".join(lemmatize))
neg_rev['lemmatized_review'] = lemmatized_review      
neg_rev
```

```python
# Concatenating the pos and neg data

com_rev = pd.concat([pos_rev, neg_rev], axis = 0).reset_index()
com_rev
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split \
(com_rev['lemmatized_review'].values, com_rev['mood'].values, test_size = 0.2, random_state = 101)
```

```python
# Creating train and test dataframes just to check
train_data = pd.DataFrame({'lemmatized_review': X_train, 'mood': y_train})
test_data = pd.DataFrame({'lemmatized_review': X_test, 'mood': y_test})
```

```python
train_data
```

```python
test_data
```

```python
vectorizer = TfidfVectorizer()
train_vectors = vectorizer.fit_transform(train_data['lemmatized_review'])
test_vectors = vectorizer.transform(test_data['lemmatized_review'])
```

```python
from sklearn import svm
from sklearn.metrics import classification_report
```

```python
classifier = svm.SVC(kernel = 'linear')
classifier.fit(train_vectors, train_data['mood'])
```

```python
pred = classifier.predict(test_vectors)
```

```python
report = classification_report(test_data['mood'], pred, output_dict = True)
print(f"positive {report['1.0']['recall']}")
print(f"negative {report['0.0']['recall']}")
```
